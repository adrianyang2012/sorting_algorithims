```python

import random
import matplotlib.pyplot as plt
import math
from timeit import default_timer as timer
def see_all(li):
  for i in range(0,len(li)):
    if not (li[0] == li[i]):
      return False
  return True
def min_sort(li):
  new_li = []
  while li:
    new_li.append(min(li))
    li.remove(min(li))
  return new_li
def quick_sort(li,see_proccess=False):
  if see_all(li):
    try:
      t = li[0]
      return li
    except:
      return []
  else:
    choose = li[0]
    li.remove(choose)
    sort_1 = [l for l in li if choose <= l]
    sort_2 = [l for l in li if choose > l]
    if see_proccess:
      print(f"list we are sorting now: {li}")
      print(f"we are spliting by {choose}")
      print(f"we are doing these to lists now: {sort_1}, {sort_2}")
    return quick_sort(sort_2)+[choose]+quick_sort(sort_1)
def remove_nan(li):
  li2 = []
  for i in range(1,len(li)):
    if li[i]:
      li2.append(li[i])
  return li2
def new_sort(li):
  try:
    li2 = [[] for i in range(0,max(li))]
    for i in range(0,len(li)):
      li2[li[i]-1].append(li[i])
    for i in range(0,len(li2))
    return li2.remove_nan
  except:
    return
```

```python
sort_1s = []
their_sorts = []
quick_sorts = []
new_sorts = []
wrongs = 0
wrongs2 = 0
marker = 1
test_min_sort = 0
test_new_sort = 1

for i in range(1,2000):
  if marker:
    if i % 100 == 0:
      print(i)
  rand_list = [random.randint(0,100) for i in range(0,i)]

  s = timer()
  y = sorted(rand_list)
  e = timer()
  their_sorts.append(e-s)
  s = timer()
  z = quick_sort(rand_list)
  e = timer()
  quick_sorts.append(e-s)
  if not z==y:
    wrongs+=1
  if test_new_sort:
    s = timer()
    n = new_sort(rand_list)
    e = timer()
    new_sorts.append(e-s)
    if not n==y:
      wrongs2+=1
  if test_min_sort:
    s = timer()
    x = min_sort(rand_list)
    e = timer()
    sort_1s.append(e-s)
if test_min_sort:
  plt.plot([math.log(sort_1) for sort_1 in sort_1s],label='min sort')
if test_new_sort:
  plt.plot([math.log(new_sort) for new_sort in new_sorts],label='new sort')
plt.plot([math.log(sorts) for sorts in their_sorts],label='quick sort')
plt.plot([math.log(quicks) for quicks in quick_sorts],label='built in sort')
plt.title('different sort algorithm\'s')
plt.legend(loc='upper left')
plt.show()
```
