```python

import random
import matplotlib.pyplot as plt
import math
from timeit import default_timer as timer
def see_all(li):
  for i in range(0,len(li)):
    if not (li[0] == li[i]):
      return False
  return True
def min_sort(li):
  new_li = []
  while li:
    new_li.append(min(li))
    li.remove(min(li))
  return new_li
def quick_sort(li,see_proccess=False):
  if see_all(li):
    try:
      t = li[0]
      return li
    except:
      return []
  else:
    choose = li[0]
    li.remove(choose)
    sort_1 = [l for l in li if choose <= l]
    sort_2 = [l for l in li if choose > l]
    if see_proccess:
      print(f"list we are sorting now: {li}")
      print(f"we are spliting by {choose}")
      print(f"we are doing these to lists now: {sort_1}, {sort_2}")
    return quick_sort(sort_2)+[choose]+quick_sort(sort_1)
```

```python
sort_1s = []
their_sorts = []
quick_sorts = []
wrongs = 0
for i in range(0,1000):
  rand_list = [random.randint(0,100) for i in range(0,i)]
  s = timer()
  x = min_sort(rand_list)
  e = timer()
  sort_1s.append(e-s)
  s = timer()
  y = sorted(rand_list)
  e = timer()
  their_sorts.append(e-s)
  s = timer()
  z = quick_sort(rand_list)
  e = timer()
  quick_sorts.append(e-s)
  if not z==y:
    wrongs+=1
plt.plot([math.log(sort_1) for sort_1 in sort_1s])
plt.plot([math.log(sorts) for sorts in their_sorts])
plt.plot([math.log(quicks) for quicks in quick_sorts])
```
